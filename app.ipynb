```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("VideoGameUsage_Profile.csv")
```

```python
df
```

```python
df.info()
```

```python
# The ratio of students

plt.figure(figsize=(8, 7))
plt.pie(df["student"].value_counts(), autopct=lambda pct: f"{round(pct, 2)}%" + "\nStudent" if pct > 80 else f"{round(pct, 2)}%" + "\nNot Student", colors=['skyblue', 'lightcoral'])
plt.title("Student Vs Not Student")
plt.show()
```

```python
# So, most of people asked in this survey are students. And that beacause the survey was done in a computer science departement
```

```python
# Male and female
genders = df["Gender"].value_counts()
plt.bar(genders.index, genders)
plt.title("Partition of Males and Females")
plt.text(0, genders[0] - 5, str(round(genders[0] / 99 * 100, 1)) + "%", ha='center', color="white")
plt.text(1, genders[1] - 3, str(round(genders[1] / 99 * 100, 1)) + "%", ha='center', color="white")
plt.show()
```

```python
# The most of people asked are males with a percentage of 92.9%, so we can't do comparaison between the two genders because we don't have enough data on Females
```

```python
# Distribution of Age
plt.figure()
plt.hist(df["Age"])
plt.title("Age Distribution")
plt.show()
```

```python
# The most common age is 23 as we can see,The max is 27 years old  and the youngest is 21
# It is obvious, this is the age of a student in university
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Age', hue='play_video_game', palette='viridis')
plt.xlabel('Age')
plt.ylabel("")
plt.title('Comparison of Video Game Playing Habits by Age')
plt.legend(title='Play Video Games', loc='upper right')
plt.yticks([])
plt.show()
```

```python
# So, As we can see all students between 21 and 26 years old are playing video games, and some of them just sometimes
# Just 27 years old, most of them just playin sometimes (Are getting old😁, I'm just joking you still young🙈)
```

```python
# Now, will check if the favorite game of a person is always his most played game
# Display where it is not the same
df[df["favorite_game"] != df["most_played_game"]]
```

```python
# So actually there is just one case for the student number 79, where his favorite game is Fortnite, but he plays COD most of his time
# Of course it can happen. Me for example: my favorite is GTA 5, but I play PES the most, because I play with friends
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(data=df, x='reason_of_play_game', hue='Age', palette='viridis')
plt.title("Age and Reson Of Playing")
plt.show()
```

```python
# Reasons Of playing Games
reason_games = df["reason_of_play_game"].value_counts()

plt.figure(figsize=(12, 5))
sns.barplot(x=reason_games.index, y=reason_games, palette="viridis")
plt.xlabel('Resaon')
plt.ylabel('Count')
plt.title('Reasons Of Playing Games')
plt.xticks(rotation=45)
plt.grid(False)
```

```python
# As excpected, Most of people play Games for entertainment (This is for what are games made), and also some for relaxation and creativity. 
# Some of them are competing
# But what is GTA Vice city doing there, either it is just a mistake or there is someone who is obsessed with this game😂
```
